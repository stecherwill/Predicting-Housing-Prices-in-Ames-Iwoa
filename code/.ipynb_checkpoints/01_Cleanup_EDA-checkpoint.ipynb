# Ames Housing Data and Kaggle Challenge

## Problem Statement

Want to build a regression model based on the Ames Housing Dataset. This model will predict the price of a house at sale as well as show what improvements can be made to increase the value of a home. In order to make this model I will creatively handle missing and null values. Use feature engineering to develop unique interaction terms using polynomial features. Then implement regularization to produce a quality cross value score and R2 score that will be the most efficient to predict the price of a house at sale.  

## Importing and Cleaning Data

- #### External Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from data_preprocessing import *

%matplotlib inline
```

- #### Importing Train and Test Housing Data

```python
# Import Train data
df_train = pd.read_csv('../data/train.csv')
df_train.head()
```

```python
# Import Test data
df_test = pd.read_csv('../data/test.csv')
```

- #### Inspecting Types of Data

```python
df_train.info()
```

- #### Filling null values for numerical columns (using functions from data preprocessing text file)

```python
impute_missing_values(df_train, 'random')
impute_missing_values(df_test, 'random')
```

- #### Grabbing all categorical features and filling missing values with string null values

```python
cat_features = df_train.select_dtypes(include = 'object').columns
cat_features

for category in cat_features:
    df_train[category].fillna('na', inplace = True)
    df_test[category].fillna('na', inplace = True)
```

- #### Converting all columns to lowercase and replacing spaces with underscores in columns.

```python
df_train.columns = [i.lower().replace(' ', '_') for i in df_train.columns]
df_test.columns = [i.lower().replace(' ', '_') for i in df_test.columns]
```

## EDA

#### Plotting Histograms for Numerical Data

```python
features = [col for col in df_train.select_dtypes(exclude='object').columns]
subplot_hist(df_train, features);
```

#### Takeaways:

- Looking into what features have a more normal distribution to help gain insights on what might best be used to predict sale price.
- Features with one straight line will not be a good predictor becuase it all falls into one bin and will not help differentiate between houses.
