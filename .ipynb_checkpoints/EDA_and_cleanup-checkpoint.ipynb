```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df_train = pd.read_csv('./datasets/train.csv')
df_train.head()
```

```python
df_train.info()
```

```python
# Filling null vales of the ints and floats
df_train = df_train.apply(lambda col: col.fillna(0) if col.dtype == 'int64' or col.dtype == 'float64' else col)
```

```python
df_train.info()
```

```python
# Convert all columns to lowercase and replace spaces in column names.
df_train.columns = [i.lower().replace(' ', '_') for i in df_train.columns]
```

```python
df_train.head()
```

```python
# looking at 
for _, col in enumerate(df_train):
    value_counts = df_train[col].value_counts()
    print(value_counts, 'count:', value_counts.sum())
```

```python
df_train_corr = df_train.corr()
df_train_corr
```
